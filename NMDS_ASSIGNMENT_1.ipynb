```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("/content/train.csv")
print(df)
```

```python
print(df.info())
print('-' * 40)
print(df.info())
```

```python
df.describe()
```

1. Overall survival rate of passengers

```python
survived_data = df[df['Survived'] == 1]
survived = survived_data.count().values[1]
survival_percent = (survived/891) * 100
print('The percentage of survived people in titanic is {}'.format(survival_percent))
```

2. Gender Distribution in titanic

```python
df.Sex.value_counts()
```

```python
df.Sex.value_counts().plot(kind='pie')
```

3. Survival rate by gender

```python
sns.catplot(x ="Sex", hue ="Survived",kind ="count", data = df)
```

4. Age distribution of passengers in titanic

```python
sns.barplot(x='Sex', y='Age', data=df)
```

5. Survival rate by journey class

```python
print(pd.pivot_table(df, index = 'Survived', columns = 'Pclass',values = 'Ticket' ,aggfunc ='count'))
print()
print(pd.pivot_table(df, index = 'Survived', columns = 'Sex', values = 'Ticket' ,aggfunc ='count'))
print()
print(pd.pivot_table(df, index = 'Survived', columns = 'Embarked',  values = 'Ticket' ,aggfunc ='count'))
```

```python
sns.catplot(x ='Embarked', hue ='Survived',kind ='count', col ='Pclass', data = df)
```
